```python
# инстансы объекта 

list1 = [] # [] - список - структура данных. Структура данных - контейнер который хранит все данные
list2 = list1 # count link 2
id(list1) == id(list2), list1 is list2  # id объекта // is проверяет принадлежность к одному объекту
del list1
print(list2)
del list2 # ничего не возвращает, нерезультирующая инструкция (void), есть ещё результирующая
# list2 объект удален из памяти, кол-во ссылок 0. Когда кол-во ссылок 0 на объект, приходит Garbage Collection и удаляет объект из памяти

#  Garbage Collection удаляет объект при кол-ве ссылок равному 0

# list1 #name 'list1' is not defined
# Любой ЯП состоит из алгоритмов и структуры данных 
```

Интерпритатор - это программа который читает и запускает код построчно, читает код сверху вниз
Интерпритатор(формальное определение) - не занимается генерацией машинного кода, он вызывает для каждой интерпретируемой команды специальную функцию 



Компилятор - это программа которая собирает весь файл программы в единое целое, и потом его запускает
Компилятор(формальное определение) - генерирует машинный код(байткод) и передает его процессору для исполнения. 

Компилятор делает работу единоразого, а интерпритатор каждый раз при чтении очередной инструкции 




Программа скомпилируется только тогда когда нет ошибок

Python - это интерпретируемый ЯП

В Python есть инструмент по очистке мусора(удаление объектов из памяти), он называется Garbage Collection
